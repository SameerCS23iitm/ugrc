```python
import pandas as pd
df = pd.read_csv("../train/labeled_1s_cscada.csv").sort_values("time_window").reset_index(drop=True)

split = int(len(df) * 0.2)
val_df  = df.iloc[:split]
test_df = df.iloc[split:]

val_df.to_csv("labeled_1s_cscada_val.csv",  index=False)
test_df.to_csv("labeled_1s_cscada_test.csv", index=False)
```

```python
print(val_df["label"].value_counts())
print(test_df["label"].value_counts())
```
