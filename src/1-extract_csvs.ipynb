```python
import subprocess
from pathlib import Path

# input_dir = "/path/to/pcap_directory"
# output_csv = "/path/to/output/all_packets.csv"

def extract (input_dir, output_csv):
    pcaps = sorted(Path(input_dir).glob("*.pcap"))
    
    cmd = [
        "tshark",
    ]
    
    # add all pcaps
    for p in pcaps:
        cmd += ["-r", str(p)]
    
    cmd += [
        "-T", "fields",
        "-e", "frame.time_epoch",
        "-e", "frame.len",
        "-e", "ip.src",
        "-e", "ip.dst",
        "-e", "modbus.func_code",
        "-e", "modbus.reference_num",
        "-e", "modbus.read_reference_num",
        "-e", "modbus.write_reference_num",
        "-e", "modbus.word_cnt",
        "-e", "modbus.read_word_cnt",
        "-e", "modbus.write_word_cnt",
        "-e", "modbus.byte_cnt",
        "-e", "modbus.exception_code",
        "-e", "modbus.response_time",
        "-e", "modbus.request_frame",
        "-e", "modbus.regnum16",
        "-e", "modbus.regval_uint16",
        "-e", "modbus.regval_float",
        "-E", "header=y",
        "-E", "separator=,",
        "-E", "quote=d",
        "-E", "occurrence=f"
    ]
    
    with open(output_csv, "w") as f:
        subprocess.run(cmd, stdout=f)
    
    print("Saved:", output_csv)
```

```python
extract("../data/benign/network-wide-pcap-capture/network-wide","../train/benign_nw.csv")
```
