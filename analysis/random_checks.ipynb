```python
import glob
import pandas as pd
from datetime import datetime, UTC

folder_path = "../train/chunks/*.csv"

distinct_values = set()

for file in glob.glob(folder_path):
    df = pd.read_csv(file, usecols=["frame.time_epoch"])
    start = df["frame.time_epoch"].min() + 10800
    end = df["frame.time_epoch"].max() + 10800
    udt_start = datetime.fromtimestamp(start, tz=UTC)
    udt_end = datetime.fromtimestamp(end, tz=UTC)
    print(file.split("/")[-1] + " | " + str(udt_start) + " | " + str(udt_end))
```
